```python
# Firecrawl connection test

import os
from dotenv import load_dotenv
from firecrawl import FirecrawlApp

# Cargar las variables del archivo .env
load_dotenv()

# Inicializar la app con la llave del entorno
app = FirecrawlApp(api_key=os.getenv("FIRECRAWL_API_KEY"))

# Probar el scrapeo de una URL simple (Sintaxis corregida para el nuevo SDK)
try:
    # En el nuevo SDK, se usa el argumento 'formats' directamente en la raíz de los parámetros de la función
    scraped_data = app.scrape_url('https://example.com', formats=['markdown'])
    
    print("✅ Conexión con Firecrawl exitosa!")
    print("\nPrimeros 100 caracteres del Markdown obtenido:")
    
    # Dependiendo de la versión exacta, la respuesta puede venir como diccionario o como un objeto con atributos
    if isinstance(scraped_data, dict):
        print(scraped_data.get('markdown', '')[:100])
    else:
        print(scraped_data.markdown[:100])
        
except Exception as e:
    print(f"❌ Error al conectar con Firecrawl: {e}")
```
